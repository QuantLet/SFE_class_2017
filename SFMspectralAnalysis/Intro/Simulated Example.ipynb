```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import scipy
```

```python
s_arr = [0.25, 0.5, 0.75]
plt.figure(figsize=(16/2,9/2))
ax0 = plt.subplot(212)
ax1 = plt.subplot(231)
ax2 = plt.subplot(232, sharey = ax1)
ax3 = plt.subplot(233, sharey = ax1)

ax_list = [ax1,ax2,ax3]
color_list = ['b','orange', 'green']
for i, s in enumerate(s_arr):
    AR1 = [0]
    x0 = 0
    a = 0.5
#     s = 0.25
    length = 1000
    for _ in range(length):
        e = np.random.normal(scale=s)
        AR1.append(a*AR1[-1]+e)

    AR1 = np.array(AR1)
    ax_list[i].plot(AR1, color=color_list[i])
    
    nu_arr = np.linspace(0,0.5)
    
    f = lambda nu: s**2/(1-2*a*np.cos(2*np.pi*nu)+a**2)# spectral density
    ax0.plot(nu_arr, f(nu_arr), color_list[i])

ax1.set_ylabel(r'$X_t$')

ax0.set_xlabel(r'$\omega$')
ax0.set_ylabel(r'$S(\omega)$')
ax0.set_ylim((0,3))
plt.savefig('AR1Per_sigma.png', transparent=True)
```
