```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy import signal
```

```python
omega1 = .75
omega2 = .6
omega3 = .5

f1 = lambda t: np.sin(t*omega1)
f2 = lambda t: np.sin(t*omega2)
f3 = lambda t: np.sin(t*omega3)

Ns = int(10e3) # Number of samples
T = 50         # Length of timeseries in time
fs = T/Ns
t_arr = np.linspace(0, T, Ns)

ts1 = f1(t_arr)
ts2 = f2(t_arr)
ts3 = f3(t_arr)

ts = ts1+ts2+ts3

plt.figure(figsize=(8,4))
plt.plot(t_arr, ts)
```

```python
f, Pxx_den = signal.periodogram(ts3, Ns)
# plt.semilogy(f, Pxx_den)
plt.plot(f, Pxx_den)
```

```python
f[np.argmax(Pxx_den)]*Ns
```
